# Parameterized Phantom Generation

```python
from xdesign import *
import matplotlib.pyplot as plt
import numpy as np
```

Generate 4 different phantoms from the same parameterized class.

```python
np.random.seed(0)
p1 = Foam(size_range=[0.05, 0.01], gap=0, porosity=1)
plot_phantom(p1)
plt.show()
```

```python
np.random.seed(0)
p2 = Foam(size_range=[0.07, 0.01], gap=0, porosity=0.75)
plot_phantom(p2)
plt.show()
```

```python
np.random.seed(0)
p3 = Foam(size_range=[0.1, 0.01], gap=0, porosity=0.5)
plot_phantom(p3)
plt.show()
```
