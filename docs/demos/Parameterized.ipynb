# Parameterized Phantom Generation

```python
from xdesign import *
import matplotlib.pyplot as plt
import numpy as np

SIZE = 500
DPI = 300
```

Generate 4 different phantoms from the same parameterized class.

```python
np.random.seed(0)
p1 = Foam(size_range=[0.05, 0.01], gap=0, porosity=1)
d1 = discrete_phantom(p1, SIZE)
plt.imshow(d1, cmap='viridis')
plt.show()
```

```python
np.random.seed(0)
p2 = Foam(size_range=[0.07, 0.01], gap=0, porosity=0.75)
d2 = discrete_phantom(p2, SIZE)
plt.imshow(d2, cmap='viridis')
plt.show()
```

```python
np.random.seed(0)
p3 = Foam(size_range=[0.1, 0.01], gap=0, porosity=0.5)
d3 = discrete_phantom(p3, SIZE)
plt.imshow(d3, cmap='viridis')
plt.show()
```

```python
np.random.seed(0)
p4 = Foam(size_range=[0.1, 0.01], gap=0.015, porosity=0.5)
d4 = discrete_phantom(p4, SIZE)
plt.imshow(d4, cmap='viridis')
plt.show()
```

```python
plt.imsave('parameterized_d1.png', d1, cmap='viridis', dpi=DPI)
plt.imsave('parameterized_d2.png', d2, cmap='viridis', dpi=DPI)
plt.imsave('parameterized_d3.png', d3, cmap='viridis', dpi=DPI)
plt.imsave('parameterized_d4.png', d4, cmap='viridis', dpi=DPI)
```
